```python
# Top 10 counts of countries
df1 = pd.DataFrame(df['country_txt'].value_counts().head(10))
df1.reset_index(inplace = True)
df1.columns=['country_name','counts']
df1
```

# Data Visualization

```python
# Barplot of 
plt.subplots(figsize=(20,10))
sns.barplot(x='country_name',y = 'counts',data = df1,palette = 'gnuplot2',ec='black',lw = 4)
plt.title('Top 10 attacked countries',fontsize=30)
plt.xlabel('country_name',fontsize=20)
plt.ylabel('counts',fontsize=20)
plt.show()
```

## Observation 
Iraq has been attacked maximum times followed by Pakistan and Afganistan.

```python
df2 = pd.DataFrame(df['gname'].value_counts().head(10).drop('Unknown'))
df2.reset_index(inplace = True)
df2.columns=['gname','counts']
df2
```

```python
plt.subplots(figsize=(20,10))
plt.xticks(rotation = 50)
sns.barplot(x='gname',y = 'counts',data = df2,palette = 'gist_gray',ec='black',lw = 4)
plt.title('Top 10 Terrorist group',fontsize=30)
plt.xlabel('gname',fontsize=20)
plt.ylabel('counts',fontsize=20)
plt.show()
```

## Observation
Taliban is the most active group.
